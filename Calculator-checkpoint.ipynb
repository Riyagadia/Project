```python
from tkinter import *
from math import sqrt
window = Tk ()
window.title ("Calculator")
window.geometry ("444x270")

def value (symbol):
    global calculation
    results= entry.get ()
    cal = str(symbol)
    entry.delete (0, "end")
    entry.insert (0, results + cal)
    

def calculation ():
    global calculation
    try :
        result = eval(entry.get())
        entry.delete (0, "end")
        entry.insert (0, str (result))
    except Exception as e:
        entry.delete(0, "end")
        entry.insert(0, "Error")
        

def do_sqrt():
    root = str (sqrt (float (entry.get())))
    entry.delete (0, "end")
    entry.insert (0, root)    

def clear_field ():
    global calculation
    calculation =""
    entry.delete (0, "end")
        
        
entry = Entry(window, width=65, borderwidth=5)
entry.grid (row = 0, column = 0, padx= 0, pady=10, columnspan= 20)



b1 = Button (window, text = "1", width = 5, command = lambda : value (1),  font = ("Arial", 21))
b1.grid (row= 1, column = 1)

b2 = Button (window, text = "2", width = 5, command = lambda : value (2), font = ("Arial", 21))
b2.grid (row= 1, column = 2)

b3 = Button (window, text = "3", width = 5, command = lambda : value (3),  font = ("Arial", 21))
b3.grid (row= 1, column = 3)

b4 = Button (window, text = "4", width = 5, command = lambda : value (4),  font = ("Arial", 21))
b4.grid (row= 2, column = 1)

b5 = Button (window, text = "5", width = 5, command = lambda : value (5),  font = ("Arial", 21))
b5.grid (row= 2, column = 2)

b6 = Button (window, text = "6", width = 5, command = lambda : value (6),  font = ("Arial", 21))
b6.grid (row= 2, column = 3)

b7 = Button (window, text = "7",  width = 5, command = lambda : value (7), font = ("Arial", 21))
b7.grid (row= 3, column = 1)

b8 = Button (window, text = "8", width = 5, command = lambda : value (8), font = ("Arial", 21))
b8.grid (row= 3, column = 2)

b9 = Button (window, text = "9", width = 5,  command = lambda : value (9), font = ("Arial", 21))
b9.grid (row= 3, column = 3)

b0 = Button (window, text = "0", width = 5, command = lambda : value (0), font = ("Arial", 21))
b0.grid (row= 4, column = 2)

bbracketopen = Button (window, text = "(", width = 5, command = lambda : value ("("), font = ("Arial", 21))
bbracketopen.grid (row= 4, column = 1)

bbracketclosed = Button (window, text = ")", width = 5, command = lambda : value (")"), font = ("Arial", 21))
bbracketclosed.grid (row= 4, column = 3)

bplus = Button (window, text = "+", width = 5, command = lambda : value ("+"), font = ("Arial", 21))
bplus.grid (row= 1, column = 4)

bminus = Button (window, text = "-", width = 5, command = lambda : value ("-"), font = ("Arial", 21))
bminus.grid (row= 2, column = 4)

bmul = Button (window, text = "*", width = 5, command = lambda : value ("*"), font = ("Arial", 21))
bmul.grid (row= 3, column = 4)

bmul = Button (window, text = "/", width = 5, command = lambda : value ("/"), font = ("Arial", 21))
bmul.grid (row= 4, column = 4)

bclear = Button (window, text = "c", width = 5, command = clear_field , font = ("Arial", 21))
bclear.grid (row= 1, column = 5)

bdeci = Button (window, text = ".", width = 5, command = lambda : value ("."), font = ("Arial", 21))
bdeci.grid (row= 2, column = 5)

bequal = Button (window, text = "=", width = 5, command = calculation , font = ("Arial", 21))
bequal.grid (row= 3, column = 5)

bsquroot = Button(window, text= "âˆš ", width = 5, command= do_sqrt,  font = ("Arial", 21))
bsquroot.grid (row= 4, column = 5)

window.mainloop ()
```
